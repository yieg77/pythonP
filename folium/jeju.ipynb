```python
ori_df = pd.read_csv('./jeju.csv', encoding='euc-kr')
ori_df.head(10)
```

```python
tmp_df1 = pd.pivot_table(ori_df, columns=['일반현황'], index=['지역'])#, values=['건수'])
tmp_df1.head(2)
```

```python
tmp_df=pd.DataFrame({"지역":ori_df['지역'], "인구":ori_df['일반현황']})
tmp_df['인구']=tmp_df['인구'].str.split('/').str.get(0).str.split('(').str.get(0).str.strip().str.split(' ').str.get(-1).str.replace('명','').astype('int')
tmp_df

df=tmp_df.groupby(by="지역").sum()
df
#df2=pd.pivot_table(df, index=['지역'],values=['인구'],aggfunc='sum')
#df2
#for i in tmp_df:
    
```

```python
geo_json = './jeju-municipalities-geo.json'
geo_jeju = json.load(open(geo_json, encoding='utf-8'))
geo_jeju
```

```python
import folium
map = folium.Map(location=[33.4314347,126.5724016], zoom_start=10)
# 자치구별 표시 => 범죄 컬럼 기준 칼라맵 사용
map.choropleth(geo_data=geo_jeju, 
               data=df['인구'], 
               columns=[df.index, df['인구']], 
               key_on='feature.id',
               fill_color='YlGn', legend_name='제주도 시별 인구수')
map
```
